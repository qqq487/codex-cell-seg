```python
import cv2
import torch
import numpy as np
from torchvision import transforms
from PIL import Image, ImageOps
```

```python
def preprocess(pil_img, is_mask):
    # w, h = pil_img.size
    # newW, newH = int(scale * w), int(scale * h)
    # assert newW > 0 and newH > 0, 'Scale is too small, resized images would have no pixel'
    # pil_img = pil_img.resize((newW, newH), resample=Image.NEAREST if is_mask else Image.BICUBIC)
    img_ndarray = np.asarray(pil_img)

    if img_ndarray.ndim == 2 and not is_mask:
        img_ndarray = img_ndarray[np.newaxis, ...]
    elif not is_mask:
        img_ndarray = img_ndarray.transpose((2, 0, 1))

    if not is_mask:
        img_ndarray = img_ndarray / 255

    return img_ndarray
```

```python
from torchvision import transforms
import torchvision.transforms.functional as TF
import random
import matplotlib.pyplot as plt
```

```python
for q in range(1):
    # Random crop
    i, j, h, w = transforms.RandomCrop.get_params(image, output_size=(300, 300))
    _image = TF.crop(image, i, j, h, w)
    print(i,j,h,w)

#     # Random horizontal flipping
#     if random.random() > 0.5:
#         _image = TF.hflip(_image)

#     # Random vertical flipping
#         _image = TF.vflip(_image)


    # Transform to tensor
    _image0 = TF.to_tensor(_image)

    mean, std = _image0.mean([1,2]), _image0.std([1,2])

    transform_norm = transforms.Compose([
        transforms.Normalize(mean, std),
        #transforms.RandomAutocontrast()
        #transforms.ColorJitter(brightness=1, contrast=1, saturation=0, hue=0)
    ])
    _image0 = transform_norm(_image0)

    # convert tis image to numpy array
    img_normalized0= np.array(_image0)

    # transpose from shape of (3,,) to shape of (,,3)
    img_normalized0 = img_normalized0.transpose(1, 2, 0)
    
    plt.imshow(img_normalized0)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("QQQ/1.png")
    
    
    # _image = TF.rotate(_image, 30, expand = False)
    _image = TF.resized_crop(image, i, j, 500, 500, size=(500, 100))
    # Transform to tensor
    _image = TF.to_tensor(_image)
    
    print(_image.size())

    mean, std = _image.mean([1,2]), _image.std([1,2])

    transform_norm = transforms.Compose([
        transforms.Normalize(mean, std),
        #transforms.RandomAutocontrast()
        #transforms.ColorJitter(brightness=1, contrast=1, saturation=0, hue=0)
    ])
    _image = transform_norm(_image)

    # convert tis image to numpy array
    img_normalized = np.array(_image)

    # transpose from shape of (3,,) to shape of (,,3)
    img_normalized = img_normalized.transpose(1, 2, 0)
    
    plt.imshow(img_normalized)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("QQQ/2.png")
```
